## **[6]**  Seleccionando Panamá realice un gráfico de barras con los valores disponibles

````python
#Seleccionar solo los datos de Panamá
datos_pma = df[df['Pais'] == 'Panama']

#Graficar los datos de Panamá
plt.figure(figsize=(12, 6))
bars = plt.bar(datos_pma['Periodo'], datos_pma['PartosAtendidos'], color='green')
plt.title('Partos atendidos en Panamá')
plt.xlabel('Año')
plt.ylabel('Cantidad de partos atendidos')
plt.xticks(rotation=45, ha='right') #Rotar y alinear las etiquetas del eje x
plt.grid(True)

#Añadir etiquetas de los valores encima de las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', color='black')

plt.show()
````

## **[9]**  Utilizando la selección anterior, genere mediante subplot un gráfico para cada país usando un mismo lienzo y 3 subplot
### Rotule adecuadamente los ejes de sus gráficas

````python
import pandas as pd
import matplotlib.pyplot as plt


# Selección de países
paises_seleccionados = ["Panama", "Denmark", "Mauritius"]

# Creación de subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Panamá
panama_df = new_df[new_df['Pais'] == "Panama"]
x_values_panama = panama_df['Periodo']
y_values_panama = panama_df['PartosAtendidos']  # Assuming 'PartosAtendidos' is the y-axis data
ax1.plot(x_values_panama, y_values_panama, marker='o', color='blue', label='Panamá')  # Agregar simbolo 'o' y color azul
ax1.set_title("Panamá")
ax1.set_xlabel('Periodo')
ax1.set_ylabel('Partos Atendidos')
ax1.legend()  # Agregar leyenda
ax1.tick_params(axis='x', rotation=45)  # Rotar etiquetas del eje x

# Dinamarca
dinamarca_df = new_df[new_df['Pais'] == "Denmark"]
x_values_dinamarca = dinamarca_df['Periodo']
y_values_dinamarca = dinamarca_df['PartosAtendidos']  
ax2.plot(x_values_dinamarca, y_values_dinamarca, marker='s', color='red', label='Dinamarca')  # Agregar simbolo's' y color verde
ax2.set_title("Dinamarca")
ax2.set_xlabel('Periodo')
ax2.set_ylabel('Partos Atendidos')
ax2.legend()  # Agregar leyenda
ax2.tick_params(axis='x', rotation=45)  # Rotar etiquetas del eje x

# Mauricio
mauritius_df = new_df[new_df['Pais'] == "Mauritius"]
x_values_mauritius = mauritius_df['Periodo']
y_values_mauritius = mauritius_df['PartosAtendidos']  
ax3.plot(x_values_mauritius, y_values_mauritius, marker='^', color='green', label='Mauricio')  # Agrega simbolo '^' y color rojo
ax3.set_title("Mauricio")
ax3.set_xlabel('Periodo')
ax3.set_ylabel('Partos Atendidos')
ax3.legend()  # Agregar leyenda
ax3.tick_params(axis='x', rotation=45)  # Rotar etiquetas del eje x

# Ajuste del layout y visualización del gráfico
fig.suptitle('Análisis de Partos Atendidos en Panamá, Dinamarca y Mauricio')
plt.tight_layout()
plt.show()
````

## Generación de un dataset que contenga solo los 20 países con mayor cantidad de datos

En este caso se utiliza la funcion *groupby* del dataframe para obtener la cantidad de valores que se tienen para cada país y al mismo tiempo la funcion *sort_values* para que este conteo se ordene de mayor a menor. 

Posteriormente se seleccionan de esta serie solo los países que tienen  $20$ o más datos/regitros en el dataset. Es decir países de los que se tienen información de la atención en al menos veinte años. 

La variable *listaCortapaises* es una serie que contiene los países y el valor contado para cada país, en esta el índice(*index*) de la serie es el nombre del país. Para realizar nuestra selección se pide a la serie que dichos índices sean convertidos a una lista. 

Luego se selecciona del dataframe completo solo los países que se encuentran en *paisesSeleccion* usando el metodo *.isin()* del dataframe de Pandas. De esta manera nuestro *dfSelecion* solo contiene los registros de los 20 países con mayor cantidad de datos.

**Nota** : En este código *dfPartos* es nuestro dataset cargado a través de Pandas, por lo cual, debe colocar el nombre que usted asignó a esta variable.

````python
# Se realiza el conteo 
# Se genera una serie de datos ordenada de acuerdo al pais con mas datos
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV y separar las columnas según el punto y coma
df = pd.read_csv('AtencionPartos.csv', sep=';')

# Realizar el conteo de la cantidad de datos por país y ordenarlos de mayor a menor
conteo = df.groupby(['Pais'])['Periodo'].count().sort_values(ascending=False)
````

````python
# Se seleccionan los primeros 20 paises
listaCortapaises = conteo[:20]  #conteo>18
````

````python
paisesSeleccion = listaCortapaises.index.tolist()
````

````python
dfSelecion = df[(df['Pais'].isin(paisesSeleccion))]
````

````python
dfSelecion.groupby('Pais').count()
````

## **[13]** Usando SubPlots genere un gráfico que contenga las 20 gráficas con el índice por año de cada país.
### Para este gráfico coloque adecuadamente las etiquetas a cada eje y el título del gráfico general y de cada gráfica individual. 
### Guarde este gráfico en un formato PDF

**Nota:** Recuerde que puede usar un ciclo for para recorrer su matriz de gráficos y de esa manera trabajar cada gráfico de manera independiente.

````python
  fig, axes = plt.subplots(5, 4, figsize=(15, 15))  

# Iteraraciones para los graficos
for i, pais in enumerate(paisesSeleccion):
    paises = dfSelecion[dfSelecion['Pais'] == pais]
    ax = axes[i // 4, i % 4]  # Calcula las coordenadas del subgráfico actual
    ax.plot(paises['Periodo'], paises['PartosAtendidos'], label=pais)
    ax.set_title(f"{pais}")
    ax.legend()
    ax.set_ylim([0, 120])  # Ajusta los límites del eje y
    ax.tick_params(axis='x', rotation=45)  # Rota las etiquetas del eje x

# Diseño
plt.tight_layout()
plt.savefig("AnDa_Lab_3_RebecaMendoza.pdf")
plt.show()
````
